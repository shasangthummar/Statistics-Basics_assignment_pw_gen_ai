**QUESTION 1: What is statistics, and why is it important?**  

**Answer:**  
Statistics is the branch of mathematics that deals with collecting, organizing, analyzing, interpreting, and presenting data. It helps us make informed decisions based on data patterns and trends.  

**Importance of Statistics:**  
1. **Data Analysis:** It allows us to interpret complex data and find meaningful insights.  
2. **Decision Making:** Businesses, governments, and researchers use statistics to make informed decisions.  
3. **Predictive Analysis:** It helps in forecasting future trends, such as weather prediction and market trends.

**QUESTION 2: What are the two main types of statistics?**  

**Answer:**  
Statistics is mainly divided into:  
1. **Descriptive Statistics:** Summarizes and describes data using measures such as mean, median, and mode.  
2. **Inferential Statistics:** Makes predictions and generalizations about a population based on a sample.  

**Example:**  
- **Descriptive Statistics:** Finding the average marks of students in a class.  
- **Inferential Statistics:** Predicting the average height of all adults in a country based on a survey sample.  

---

**QUESTION 3: What are descriptive statistics?**  

**Answer:**  
Descriptive statistics help summarize and describe the main features of a dataset. It does not draw conclusions beyond the given data.  

**Types of Descriptive Statistics:**  
1. **Measures of Central Tendency:** Mean, Median, and Mode.  
2. **Measures of Dispersion:** Range, Variance, and Standard Deviation.  
3. **Data Distribution:** Histograms, Frequency Tables, and Graphs.  

**Example:**  
- The average temperature of a city over a week is 30°C. This is a descriptive statistic as it summarizes data without making predictions.  

---

**QUESTION 4: What is inferential statistics?**  

**Answer:**  
Inferential statistics is used to make predictions or inferences about a population based on a sample. It helps in hypothesis testing and estimating probabilities.  

**Techniques Used in Inferential Statistics:**  
1. **Hypothesis Testing:** Determines if there is enough evidence to support a claim.  
2. **Confidence Intervals:** Estimates the range in which a population parameter is likely to fall.  
3. **Regression Analysis:** Predicts relationships between variables.  

**Example:**  
- A survey of 1,000 people is used to predict election results for an entire country.

**QUESTION 5: What is sampling in statistics?**  

**Answer:**  
Sampling is the process of selecting a subset of individuals from a population to analyze and make conclusions about the entire population. It is used when collecting data from the entire population is impractical.  

**Types of Sampling:**  
1. **Random Sampling:** Every individual has an equal chance of being selected.  
2. **Stratified Sampling:** The population is divided into groups, and samples are taken from each.  
3. **Cluster Sampling:** The population is divided into clusters, and entire clusters are selected randomly.  
4. **Systematic Sampling:** Every nth individual is selected from a list.  

**Example:**  
- Selecting 500 students from a university to analyze student performance instead of surveying all students.  

---



**QUESTION 6: What are the different types of sampling methods?**  

**Answer:**  
Sampling methods are techniques used to select a subset of individuals from a population for statistical analysis. They are categorized into **probability** and **non-probability sampling**.  

**1. Probability Sampling (Each individual has an equal chance of selection)**  
- **Simple Random Sampling:** Every individual has an equal chance of being selected (e.g., lottery draw).  
- **Stratified Sampling:** The population is divided into subgroups (strata), and samples are taken from each (e.g., dividing students by grade and selecting from each grade).  
- **Cluster Sampling:** The population is divided into groups (clusters), and entire clusters are randomly selected (e.g., selecting random schools instead of individual students).  
- **Systematic Sampling:** Every nth individual is chosen from a list (e.g., selecting every 10th customer in a queue).  

**2. Non-Probability Sampling (Not all individuals have an equal chance of selection)**  
- **Convenience Sampling:** Selecting individuals who are easiest to reach (e.g., interviewing people at a mall).  
- **Judgmental Sampling:** Selecting individuals based on expertise or judgment.  
- **Quota Sampling:** Selecting individuals to meet a fixed quota (e.g., 50 males and 50 females).  
- **Snowball Sampling:** Existing subjects recruit more subjects (used in hidden populations like drug users).

The main difference between **random sampling** and **non-random sampling** is how individuals are selected from the population.  

| Feature         | Random Sampling | Non-Random Sampling |
|---------------|----------------|------------------|

**Example:**

   - Labels or names without a meaningful order.

4. **Ratio Data (Quantitative, True Zero Exists)**

**QUESTION 11: What is the measure of central tendency?**  

**Answer:**  
The **measure of central tendency** is a statistical concept used to find the center or typical value of a dataset. It helps summarize large datasets with a single value representing the data's central location.  

The three main measures of central tendency are:  
1. **Mean (Average)** – The sum of all values divided by the number of values.  
2. **Median** – The middle value when the data is arranged in ascending order.  
3. **Mode** – The most frequently occurring value in the dataset.  

---

**QUESTION 12: Define mean, median, and mode.**  

**Answer:**  

1. **Mean (Arithmetic Average):**  
   - Formula:  
     \[
     \text{Mean} = \frac{\sum X}{N}
     \]  
     where \(X\) represents all data points, and \(N\) is the number of data points.  
   - **Example:**  
     - Data: 5, 10, 15, 20  
     - Mean = (5 + 10 + 15 + 20) / 4 = **12.5**  

2. **Median (Middle Value):**  
   - The middle number when data is arranged in ascending order.  
   - **For odd number of elements**: The middle value.  
   - **For even number of elements**: The average of the two middle values.  
   - **Example:**  
     - Data: 3, 5, 7, 9, 11 → Median = **7**  
     - Data: 3, 5, 7, 9 → Median = (5 + 7) / 2 = **6**  

3. **Mode (Most Frequent Value):**  
   - The value that appears most often in the dataset.  
   - A dataset can have **no mode, one mode, or multiple modes**.  
   - **Example:**  
     - Data: 2, 3, 3, 4, 5 → Mode = **3**  
     - Data: 1, 2, 2, 3, 3, 4 → Modes = **2 and 3** (Bimodal)

**QUESTION 14: What is variance, and how is it calculated?**  

**Answer:**  
**Variance** measures how spread out data points are from the **mean**. A higher variance indicates more variability, while a lower variance indicates that data points are closer to the mean.  

### **Formula for Variance**  
For a **population**:  
\[
\sigma^2 = \frac{\sum (X - \mu)^2}{N}
\]  
For a **sample**:  
\[
s^2 = \frac{\sum (X - \bar{X})^2}{n-1}
\]  
where:  
- \(X\) = each data point  
- \(\mu\) = population mean  
- \(\bar{X}\) = sample mean  
- \(N\) = number of population data points  
- \(n\) = number of sample data points  

### **Example Calculation:**  
Data: 2, 4, 6, 8  
Mean = (2+4+6+8)/4 = **5**  

\[
\sigma^2 = \frac{(2-5)^2 + (4-5)^2 + (6-5)^2 + (8-5)^2}{4}
\]  
\[
= \frac{9 + 1 + 1 + 9}{4} = \frac{20}{4} = **5**
\]  

Variance = **5**  

---

**QUESTION 15: What is standard deviation, and why is it important?**  

**Answer:**  
**Standard Deviation (SD)** measures how much data points deviate from the mean. It is the **square root** of variance.  

### **Formula for Standard Deviation:**  
\[
\sigma = \sqrt{\sigma^2}
\]
\[
s = \sqrt{s^2}
\]

where:  
- \(\sigma\) = population standard deviation  
- \(s\) = sample standard deviation  

### **Example:**  
Using variance **5** from the previous example,  
\[
\sigma = \sqrt{5} \approx 2.24
\]  
Standard Deviation = **2.24**  

### **Importance of Standard Deviation:**  
1. **Indicates Data Spread:** Higher SD means data is more spread out; lower SD means data is tightly clustered around the mean.  
2. **Used in Risk Analysis:** In finance and investment, SD helps measure volatility.  
3. **Helps Identify Outliers:** Large SD suggests the presence of extreme values.  
4. **Essential in Normal Distribution:** In a normal distribution, **68% of data** falls within **1 SD**, **95% within 2 SDs**, and **99.7% within 3 SDs**.

**QUESTION 16: Define and explain the term range in statistics.**  

**Answer:**  
**Range** is the **simplest measure of dispersion** that shows the **difference between the highest and lowest values** in a dataset.  

### **Formula:**  
\[
\text{Range} = \text{Maximum Value} - \text{Minimum Value}
\]

### **Example:**  
- Data: **3, 7, 12, 18, 25**  
- **Range** = 25 - 3 = **22**  

### **Significance of Range:**  
1. **Gives a Quick Overview:** It provides a rough idea of how spread out the values are.  
2. **Easy to Compute:** Simple to calculate but does not consider all data points.  
3. **Affected by Outliers:** If one extreme value is very high or low, the range can be misleading.

**QUESTION 17: What is the difference between variance and standard deviation?**  

**Answer:**  

| Feature           | Variance (\(\sigma^2\) or \(s^2\)) | Standard Deviation (\(\sigma\) or \(s\)) |
|------------------|---------------------------------|----------------------------------|
| **Definition**   | The average of the squared differences from the mean. | The square root of variance. |
| **Formula** (Population) | \(\sigma^2 = \frac{\sum (X - \mu)^2}{N}\) | \(\sigma = \sqrt{\sigma^2}\) |
| **Formula** (Sample) | \(s^2 = \frac{\sum (X - \bar{X})^2}{n-1}\) | \(s = \sqrt{s^2}\) |
| **Unit**        | Squared unit of the original data. | Same unit as the original data. |
| **Interpretation** | Measures the overall spread of data values. | Measures the average deviation of data from the mean. |
| **Example** (Data: 2, 4, 6, 8) | Variance = **5** | Standard Deviation = **√5 ≈ 2.24** |

### **Key Difference:**  
- **Variance** is harder to interpret because it’s in squared units.  
- **Standard deviation** is easier to interpret because it's in the same units as the data.

**QUESTION 18: What is skewness in a dataset?**  

**Answer:**  
**Skewness** measures the **asymmetry** of the distribution of data.  

- If data is **symmetrical**, skewness is **zero**.  
- If data is **not symmetrical**, it is either **positively or negatively skewed**.  

### **Types of Skewness:**  
1. **Positive Skewness (Right-Skewed)**:  
   - **Tail is longer on the right.**  
   - **Mean > Median > Mode**  
   - Example: Income distribution, where a few people earn extremely high salaries.  

2. **Negative Skewness (Left-Skewed)**:  
   - **Tail is longer on the left.**  
   - **Mean < Median < Mode**  
   - Example: Test scores where most students score high, but a few score very low.  

### **Formula for Skewness:**  
Pearson’s coefficient of skewness:  
\[
\text{Skewness} = \frac{3 (\text{Mean} - \text{Median})}{\text{Standard Deviation}}
\]

---

**QUESTION 19: What does it mean if a dataset is positively or negatively skewed?**  

**Answer:**  

| Skewness Type  | Shape | Mean, Median, Mode Relationship | Example |
|---------------|-------|---------------------------------|---------|
| **Positive Skew (Right-Skewed)** | Tail on the **right** | **Mean > Median > Mode** | Salaries, housing prices |
| **Negative Skew (Left-Skewed)** | Tail on the **left** | **Mean < Median < Mode** | Test scores, lifespan of electronic devices |
| **Zero Skew (Symmetrical)** | Bell-shaped | **Mean ≈ Median ≈ Mode** | Heights of people, IQ scores |

### **How to Interpret Skewness?**  
- **Highly Positive (>1):** Strong right-skew.  
- **Slightly Positive (0.5 to 1):** Moderate right-skew.  
- **Near Zero (-0.5 to 0.5):** Almost symmetrical.  
- **Slightly Negative (-1 to -0.5):** Moderate left-skew.  
- **Highly Negative (< -1):** Strong left-skew.

**QUESTION 20: Define and explain kurtosis.**  

**Answer:**  
**Kurtosis** measures whether the data has **heavy tails** (outliers) or **light tails** compared to a normal distribution.  

### **Types of Kurtosis:**  
1. **Mesokurtic (\(K = 3\)):**  
   - **Normal distribution** (bell-shaped).  
   - Example: Heights of people.  

2. **Leptokurtic (\(K > 3\)):**  
   - **High peak, heavy tails** (more extreme values).  
   - Example: Stock market returns (large fluctuations).  

3. **Platykurtic (\(K < 3\)):**  
   - **Flatter peak, light tails** (fewer extreme values).  
   - Example: Test scores with a narrow range.  

### **Formula for Kurtosis:**  
\[
K = \frac{\sum (X - \bar{X})^4}{(N-1) s^4}
\]
where:  
- \(X\) = each data point  
- \(\bar{X}\) = mean  
- \(s\) = standard deviation

**QUESTION 21: What is the purpose of covariance?**  

**Answer:**  
Covariance measures the **relationship between two variables** and indicates whether they **increase or decrease together**.  

### **Formula:**  
For population covariance:  
\[
\text{Cov}(X, Y) = \frac{\sum (X_i - \mu_X)(Y_i - \mu_Y)}{N}
\]  
For sample covariance:  
\[
\text{Cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
\]  

### **Interpretation of Covariance:**  
1. **Positive Covariance (\(+\))**: Both variables increase together.  
   - Example: Higher study time leads to higher exam scores.  
2. **Negative Covariance (\(-\))**: One variable increases, the other decreases.  
   - Example: More exercise leads to lower body weight.  
3. **Zero Covariance (\(0\))**: No relationship between the variables.  

⚠ **Covariance does not indicate strength, only direction!**  

---

**QUESTION 22: What does correlation measure in statistics?**  

**Answer:**  
**Correlation** measures both the **direction and strength** of the relationship between two variables. It is the **normalized version of covariance**.  

### **Formula:**  
\[
r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
\]  

where:  
- \(r\) = correlation coefficient  
- \(\sigma_X\), \(\sigma_Y\) = standard deviations of \(X\) and \(Y\)  

### **Types of Correlation:**  
| Correlation Type | Value of \(r\) | Meaning |
|-----------------|---------------|---------|
| **Perfect Positive** | \(+1\) | Variables increase together perfectly. |
| **Strong Positive** | \(+0.7\) to \(+1\) | Strong relationship. |
| **Weak Positive** | \(0\) to \(+0.7\) | Weak relationship. |
| **No Correlation** | \(0\) | No relationship. |
| **Weak Negative** | \(-0.7\) to \(0\) | Weak inverse relationship. |
| **Strong Negative** | \(-1\) to \(-0.7\) | Strong inverse relationship. |
| **Perfect Negative** | \(-1\) | Variables decrease together perfectly. |

---

**QUESTION 23: What is the difference between covariance and correlation?**  

**Answer:**  

| Feature           | Covariance  | Correlation |
|------------------|------------|-------------|
| **Definition**   | Measures how two variables change together. | Measures the strength and direction of the relationship. |
| **Formula**      | \(\text{Cov}(X, Y) = \frac{\sum (X - \bar{X})(Y - \bar{Y})}{N}\) | \(r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}\) |
| **Scale**        | Unbounded (values can be very large or very small). | Always between \(-1\) and \(+1\). |
| **Interpretation** | Only tells **direction**, not strength. | Tells **both direction and strength**. |
| **Units**        | Original units of \(X\) and \(Y\). | Unitless (dimensionless). |

💡 **Example:**  
- **Covariance = +2000**: We know the relationship is positive, but we don’t know how strong it is.  
- **Correlation = 0.85**: We know there is a strong positive relationship.  

🔹 **Conclusion:**  
- Use **covariance** to understand the **direction** of change.  
- Use **correlation** to understand the **strength** of the relationship.

```python
# PRACTICAL QUESTION 1: How do you calculate the mean, median, and mode of a dataset using Pandas and NumPy?

import numpy as np
import pandas as pd
from scipy import stats  # For mode calculation

# Sample dataset
data = [12, 15, 14, 10, 18, 22, 19, 25, 21, 17, 14, 18, 14]

# Convert to Pandas Series
df = pd.Series(data)

# Calculate Mean, Median, and Mode
mean_value = df.mean()  # Mean using Pandas
median_value = df.median()  # Median using Pandas
mode_value = df.mode()  # Mode using Pandas (returns a Series)

# Alternative calculation using NumPy
mean_np = np.mean(data)
median_np = np.median(data)
mode_np = stats.mode(data, keepdims=True)[0]  # Mode using SciPy

# Display results
print("Mean (Pandas):", mean_value)
print("Median (Pandas):", median_value)
print("Mode (Pandas):", mode_value.tolist())  # Convert Series to list

print("Mean (NumPy):", mean_np)
print("Median (NumPy):", median_np)
print("Mode (SciPy):", mode_np.tolist())  # Convert NumPy array to list
```

```python
# PRACTICAL QUESTION 2: Write a Python program to compute the variance and standard deviation of a dataset using NumPy and Pandas.

import numpy as np
import pandas as pd

# Sample dataset
data = [12, 15, 14, 10, 18, 22, 19, 25, 21, 17]

# Convert data to Pandas Series
df = pd.Series(data)

# Calculate Variance and Standard Deviation
variance = df.var()  # Pandas method
std_dev = df.std()    # Pandas method

# Alternative calculation using NumPy
variance_np = np.var(data, ddof=1)  # ddof=1 for sample variance
std_dev_np = np.std(data, ddof=1)   # ddof=1 for sample standard deviation

# Display results
print("Variance (Pandas):", variance)
print("Standard Deviation (Pandas):", std_dev)
print("Variance (NumPy):", variance_np)
print("Standard Deviation (NumPy):", std_dev_np)
```

```python
# PRACTICAL QUESTION 4: Implement sampling techniques like random sampling and stratified sampling.

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Creating a sample dataset
data = {
    "ID": range(1, 21),
    "Age": np.random.randint(18, 60, 20),
    "Income": np.random.randint(30000, 100000, 20),
    "Category": np.random.choice(["A", "B", "C"], 20)  # Categorical variable for stratified sampling
}

# Convert to Pandas DataFrame
df = pd.DataFrame(data)

# RANDOM SAMPLING
random_sample = df.sample(n=5, random_state=42)  # Select 5 random samples
print("Random Sampling:\n", random_sample)

# STRATIFIED SAMPLING (ensuring balanced categories)
stratified_sample, _ = train_test_split(df, test_size=0.75, stratify=df["Category"], random_state=42)
print("\nStratified Sampling:\n", stratified_sample)

"""
### Explanation:
- **Random Sampling**: Selects data randomly from the dataset without considering categories.
- **Stratified Sampling**: Ensures equal representation of each category by selecting samples in proportion.
"""
```

```python
# PRACTICAL QUESTION 5: Write a Python function to calculate the range of a dataset.

import pandas as pd
import numpy as np

# Function to calculate the range of a dataset
def calculate_range(series):
    return series.max() - series.min()

# Creating a sample dataset
data = {
    "Age": np.random.randint(18, 60, 15),
    "Income": np.random.randint(30000, 100000, 15)
}

# Convert to Pandas DataFrame
df = pd.DataFrame(data)

# Calculating range for each column
age_range = calculate_range(df["Age"])
income_range = calculate_range(df["Income"])

print("Dataset:\n", df)
print("\nRange of Age:", age_range)
print("Range of Income:", income_range)

"""
### Explanation:
- **Range** is calculated as the difference between the maximum and minimum values in a dataset.
- The function `calculate_range()` returns `max - min` for a given series.
"""
```

```python
# PRACTICAL QUESTION 7: Calculate skewness and kurtosis of a dataset using Python libraries.

import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a dataset with skewness
np.random.seed(42)
data = np.concatenate([np.random.normal(50, 10, 500), np.random.normal(80, 5, 100)])

# Creating a DataFrame
df = pd.DataFrame({"Values": data})

# Calculating skewness and kurtosis
skewness = stats.skew(df["Values"])
kurtosis = stats.kurtosis(df["Values"], fisher=True)  # Fisher=True gives excess kurtosis

# Displaying results
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")

# Plot histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(df["Values"], bins=30, kde=True, color="blue")
plt.axvline(df["Values"].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(df["Values"].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.title("Histogram with Skewness & Kurtosis")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

"""
### Explanation:
- We use `scipy.stats.skew()` to calculate skewness, which measures asymmetry.
- We use `scipy.stats.kurtosis()` to calculate kurtosis, which measures the sharpness of the peak.
- The histogram visualizes the data distribution along with the mean and median.
"""
```

```python
# PRACTICAL QUESTION 8: Generate a dataset and demonstrate positive and negative skewness.

import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generating a positively skewed dataset (right-skewed)
np.random.seed(42)
positive_skew = np.random.exponential(scale=2, size=1000)

# Generating a negatively skewed dataset (left-skewed)
negative_skew = -np.random.exponential(scale=2, size=1000) + 10

# Creating a DataFrame
df = pd.DataFrame({"Positive Skew": positive_skew, "Negative Skew": negative_skew})

# Calculating skewness
pos_skewness = stats.skew(df["Positive Skew"])
neg_skewness = stats.skew(df["Negative Skew"])

# Displaying results
print(f"Skewness of Positively Skewed Data: {pos_skewness}")
print(f"Skewness of Negatively Skewed Data: {neg_skewness}")

# Plotting histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["Positive Skew"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Positively Skewed Distribution")
axes[0].axvline(df["Positive Skew"].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[0].legend()

sns.histplot(df["Negative Skew"], bins=30, kde=True, color="orange", ax=axes[1])
axes[1].set_title("Negatively Skewed Distribution")
axes[1].axvline(df["Negative Skew"].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[1].legend()

plt.show()

"""
### Explanation:
- `np.random.exponential()` generates right-skewed (positively skewed) data.
- Negative skew is created by inverting an exponential distribution.
- We use `scipy.stats.skew()` to measure skewness.
- The histograms visualize the distribution, showing how the mean shifts for each skew.
"""
```

```python
# PRACTICAL QUESTION 9: Write a Python script to calculate covariance between two datasets.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating two datasets
np.random.seed(42)
x = np.random.normal(50, 10, 100)  # Dataset 1: Mean=50, Std=10, Size=100
y = 2 * x + np.random.normal(0, 5, 100)  # Dataset 2: Correlated with x

# Creating a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Calculating covariance
cov_matrix = np.cov(df['X'], df['Y'])
cov_value = cov_matrix[0, 1]

# Displaying results
print("Covariance Matrix:\n", cov_matrix)
print(f"Covariance between X and Y: {cov_value}")

# Visualizing relationship using a scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["X"], y=df["Y"], color='blue', alpha=0.7)
plt.title("Scatter Plot of X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

"""
### Explanation:
- `np.cov(x, y)` calculates the covariance matrix.
- The covariance value (off-diagonal) shows the relationship strength.
- A positive covariance means X and Y move together.
- Scatter plot visualizes the relationship.
"""
```

```python
# PRACTICAL QUESTION 10: Write a Python script to calculate the correlation coefficient between two datasets.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating two correlated datasets
np.random.seed(42)
x = np.random.normal(50, 10, 100)  # Dataset 1
y = 2 * x + np.random.normal(0, 5, 100)  # Dataset 2 correlated with x

# Creating a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Calculating correlation coefficient using NumPy
corr_coeff_numpy = np.corrcoef(df["X"], df["Y"])[0, 1]

# Calculating correlation using Pandas
corr_coeff_pandas = df.corr().iloc[0, 1]

# Displaying results
print(f"Correlation Coefficient (NumPy): {corr_coeff_numpy}")
print(f"Correlation Coefficient (Pandas): {corr_coeff_pandas}")

# Visualizing correlation using a scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(x=df["X"], y=df["Y"], scatter_kws={'color': 'blue', 'alpha': 0.7}, line_kws={'color': 'red'})
plt.title("Scatter Plot with Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

"""
### Explanation:
- `np.corrcoef(x, y)[0,1]` computes Pearson correlation coefficient.
- `df.corr()` calculates correlation for the entire DataFrame.
- Correlation ranges from -1 to 1:
  - 1 → Perfect positive correlation
  - -1 → Perfect negative correlation
  - 0 → No correlation
- Scatter plot with regression line visualizes the relationship.
"""
```

```python
# PRACTICAL QUESTION 12: Implement and compare simple random sampling and systematic sampling.

import numpy as np
import pandas as pd

# Creating a dataset
np.random.seed(42)
data = {'ID': range(1, 101), 'Value': np.random.randint(1, 100, 100)}
df = pd.DataFrame(data)

# Simple Random Sampling (selecting 10 random samples)
random_sample = df.sample(n=10, random_state=42)

# Systematic Sampling (selecting every 10th element)
step = 10
systematic_sample = df.iloc[::step, :]

# Display results
print("Simple Random Sample:")
print(random_sample)

print("\nSystematic Sample:")
print(systematic_sample)

"""
### Explanation:
1. **Simple Random Sampling:** Selects 10 random records from the dataset using `.sample(n)`.
2. **Systematic Sampling:** Selects every 10th record from the dataset using slicing `df.iloc[::step]`.
3. This allows us to compare the randomness of simple random sampling with the structured selection of systematic sampling.
"""
```

```python
# PRACTICAL QUESTION 13: Calculate the mean, median, and mode of grouped data.

import numpy as np
import pandas as pd
from scipy import stats

# Creating a grouped dataset (Example: Student scores grouped into intervals)
data = {'Score_Range': ['0-20', '21-40', '41-60', '61-80', '81-100'],
        'Frequency': [5, 15, 25, 35, 20]}  # Frequency of occurrences in each range

df = pd.DataFrame(data)

# Finding the midpoint (class mark) for each interval
df['Midpoint'] = [(int(x.split('-')[0]) + int(x.split('-')[1])) / 2 for x in df['Score_Range']]

# Computing mean, median, and mode using grouped data formulas

# Mean = Σ(fi * xi) / Σ(fi)
mean = (df['Midpoint'] * df['Frequency']).sum() / df['Frequency'].sum()

# Median Calculation
cumulative_frequency = df['Frequency'].cumsum()
median_class_index = cumulative_frequency[cumulative_frequency >= cumulative_frequency.max() / 2].index[0]
L = int(df.loc[median_class_index, 'Score_Range'].split('-')[0])  # Lower boundary of median class
F = cumulative_frequency[median_class_index - 1] if median_class_index > 0 else 0  # Cumulative freq before median class
f = df.loc[median_class_index, 'Frequency']  # Frequency of median class
h = int(df.loc[median_class_index, 'Score_Range'].split('-')[1]) - L  # Class width

median = L + ((cumulative_frequency.max() / 2 - F) / f) * h

# Mode Calculation
mode_class_index = df['Frequency'].idxmax()
L_mode = int(df.loc[mode_class_index, 'Score_Range'].split('-')[0])
f1 = df.loc[mode_class_index, 'Frequency']
f0 = df.loc[mode_class_index - 1, 'Frequency'] if mode_class_index > 0 else 0
f2 = df.loc[mode_class_index + 1, 'Frequency'] if mode_class_index < len(df) - 1 else 0
mode = L_mode + ((f1 - f0) / ((2 * f1) - f0 - f2)) * h

# Display results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")

"""
### Explanation:
1. **Mean:** Uses weighted average formula with midpoints and frequencies.
2. **Median:** Finds the median class and applies formula to estimate median.
3. **Mode:** Uses the formula for mode in grouped data by considering neighboring frequencies.

This approach provides a statistical summary of the grouped dataset.
"""
```

```python
# PRACTICAL QUESTION 14: Simulate data using Python and calculate its central tendency and dispersion.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulating a dataset of student scores (1000 values between 0 and 100)
np.random.seed(42)  # For reproducibility
data = np.random.randint(0, 101, 1000)

# Converting to DataFrame
df = pd.DataFrame(data, columns=['Scores'])

# Calculating central tendency measures
mean = df['Scores'].mean()
median = df['Scores'].median()
mode = df['Scores'].mode().values[0]  # Mode might return multiple values, taking the first one

# Calculating dispersion measures
variance = df['Scores'].var()
std_dev = df['Scores'].std()
range_val = df['Scores'].max() - df['Scores'].min()
iqr = df['Scores'].quantile(0.75) - df['Scores'].quantile(0.25)

# Displaying the results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {range_val}")
print(f"Interquartile Range (IQR): {iqr:.2f}")

# Visualizing distribution using a histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['Scores'], bins=20, kde=True, color='blue')
plt.axvline(mean, color='red', linestyle='dashed', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', label=f'Median: {median:.2f}')
plt.axvline(mode, color='orange', linestyle='dashed', label=f'Mode: {mode}')
plt.legend()
plt.title("Distribution of Scores with Mean, Median, and Mode")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()

"""
### Explanation:
1. **Mean, Median, and Mode** represent central tendency.
2. **Variance and Standard Deviation** show data spread.
3. **Range and IQR** measure dispersion.
4. **Histogram** visualizes data distribution with KDE (Kernel Density Estimation).
"""
```

```python
# PRACTICAL QUESTION 17: Calculate the interquartile range (IQR) of a dataset.

import numpy as np
import pandas as pd

# Generate a dataset of random values
np.random.seed(42)
data = np.random.randint(10, 100, 50)  # 50 random numbers between 10 and 100

# Convert to Pandas DataFrame
df = pd.DataFrame(data, columns=['Values'])

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Values'].quantile(0.25)
Q3 = df['Values'].quantile(0.75)

# Compute the Interquartile Range (IQR)
IQR = Q3 - Q1

# Print results
print(f"Dataset:\n{df['Values'].tolist()}")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")

"""
### Explanation:
- The interquartile range (IQR) is the difference between Q3 (75th percentile) and Q1 (25th percentile).
- It measures the spread of the middle 50% of the dataset.
- Outliers are typically values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.
"""
```

```python
# PRACTICAL QUESTION 18: Implement Z-score normalization and explain its significance.

import numpy as np
import pandas as pd
import scipy.stats as stats

# Generate a dataset of random values
np.random.seed(42)
data = np.random.randint(50, 150, 20)  # 20 random numbers between 50 and 150

# Convert to Pandas DataFrame
df = pd.DataFrame(data, columns=['Original Values'])

# Compute Z-scores
df['Z-score'] = stats.zscore(df['Original Values'])

# Print results
print(df)

"""
### Explanation:
- Z-score (or standard score) measures how many standard deviations a data point is from the mean.
- Formula: Z = (X - mean) / std_dev
- A Z-score of 0 means the value is equal to the mean.
- A Z-score > 0 means the value is above the mean.
- A Z-score < 0 means the value is below the mean.
- This normalization method is useful for standardizing data for machine learning and statistical analysis.
"""
```

```python
# PRACTICAL QUESTION 24: Implement the Pearson and Spearman correlation coefficients for a dataset.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

# Generate synthetic dataset
np.random.seed(42)
size = 100

# Two correlated variables (linear relationship)
x = np.random.rand(size) * 100
y = 2.5 * x + np.random.randn(size) * 10  # Adding some noise

# Two variables with a nonlinear monotonic relationship
z = np.log(x + 1) * 20 + np.random.randn(size) * 2  # Logarithmic transformation

# Convert to DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Calculate Pearson and Spearman correlation
pearson_corr, _ = pearsonr(df['X'], df['Y'])
spearman_corr, _ = spearmanr(df['X'], df['Z'])

# Visualizing correlation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=df['X'], y=df['Y'])
plt.title(f"Pearson Correlation: {pearson_corr:.2f}")

plt.subplot(1, 2, 2)
sns.scatterplot(x=df['X'], y=df['Z'])
plt.title(f"Spearman Correlation: {spearman_corr:.2f}")

plt.tight_layout()
plt.show()

# Display correlation values
print("Pearson Correlation (X and Y - Linear):", pearson_corr)
print("Spearman Correlation (X and Z - Monotonic):", spearman_corr)

"""
### Explanation:
- **Pearson Correlation**: Measures linear relationships between two continuous variables.
- **Spearman Correlation**: Measures the monotonic relationship (increasing/decreasing trend), even if nonlinear.
- The first scatter plot visualizes a linear correlation, where Pearson is most effective.
- The second scatter plot shows a monotonic (nonlinear) trend, where Spearman is more relevant.
"""
```
